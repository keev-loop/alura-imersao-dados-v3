```python
# Compute the correlation matrix
corr = dados.loc[:,'g-0':'g-50'].corr()
```

```python
import numpy as np

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

xEntrada = dados_combinados.select_dtypes('float64')
ySaida = dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(xEntrada, ySaida, test_size = 0.2, stratify=ySaida, random_state=999)
```

```python
modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
from sklearn.tree import DecisionTreeClassifier

xEntrada = dados_combinados.select_dtypes('float64')
ySaida = dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(xEntrada, ySaida, test_size = 0.2, stratify=ySaida, random_state=376)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize=(15,10), facecolor='k')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names=x_treino.columns,
               class_names=['NÃ£o Ativado', 'Ativado'])
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

xEntrada = dados_combinados.select_dtypes('float64')
ySaida = dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(xEntrada, ySaida, test_size = 0.2, stratify=ySaida, random_state=376)

teste=[]
treino=[]

for i in range(1,15):
  modelo_arvore = DecisionTreeClassifier(max_depth = 1)
  modelo_arvore.fit(x_treino, y_treino)
  teste.append(modelo_arvore.score(x_teste, y_teste))
  treino.append(modelo_arvore.score(x_treino, y_treino))
```

```python
sns.lineplot(x=range(1,15), y=teste, label='teste')
sns.lineplot(x=range(1,15), y=treino, label='treino')
```
